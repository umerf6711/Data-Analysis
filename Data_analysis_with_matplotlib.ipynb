# Subplots function

```python
fig,a = plt.subplots(2,2)
x = np.arange(1,5)
a[0][0].plot(x,x*x)
a[0][0].set_title('square')
a[0][1].plot(x,np.sqrt(x))
a[0][1].set_title('sqrt')
a[1][0].plot(x,np.exp(x))
a[1][0].set_title('exp')
a[1][1].plot(x,np.log10(x))
a[1][1].set_title('log')
plt.show()
```

```python
#different sizes

#subplot2grid

a1 = plt.subplot2grid((3,3),(0,0),colspan=2)
a2 = plt.subplot2grid((3,3),(0,2),rowspan=3)
a3 = plt.subplot2grid((3,3),(1,0),rowspan=2,colspan=2)
x = np.arange(1,10)
a2.plot(x,np.sqrt(x))
a2.set_title('sqrt')
a1.plot(x,np.exp(x))
a1.set_title('exp')
a3.plot(x,np.log10(x))
a3.set_title('log')
plt.tight_layout()
plt.show()
```

# Types of charts

**pie chart**

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
langs = ['C','C++','C#','JAVA','python']
stu = [23,17,35,28,20]
ax.pie(stu,labels=langs,autopct='%1.2f%%')
plt.show()
```

**scatter plot**

```python
g = [100,45,66,88,90,56,87,90,100,90]
b = [20,60,78,56,54,84,90,100,70,50]
grd = np.arange(1,100,10)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.scatter(grd,g,color='r')
ax.scatter(grd,b,color='b')
ax.set_xlabel('Grades Range')
ax.set_ylabel('Grades score')
ax.set_title('scatter plot')
plt.show()
```

**contour plot**

```python
x = np.linspace(-3,3,100)
y = np.linspace(-3,3,100)
X,Y = np.meshgrid(x,y)
Z= np.sqrt(X**2 + Y**2)
fig,ax=plt.subplots(1,1)

cp = ax.contour(X,Y,Z)
fig.colorbar(cp)
ax.set_title('Contour')
plt.show()
```

```python
#quiverplot

Z= X*np.exp(-X**2 - Y**2)
v,u=np.gradient(Z,100,100)
fig,ax=plt.subplots()

q = ax.quiver(X,Y,u,v)
plt.show()
```

```python
#Violin plot

np.random.seed(10)
c1 = np.random.normal(100,10,200)
c2 = np.random.normal(80,30,200)
c3 = np.random.normal(90,20,200)
c4 = np.random.normal(70,25,200)

data_to_plot = [c1,c2,c3,c4]
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
bp = ax.violinplot(data_to_plot)
plt.show()
```

# Two dimensional plot

```python
x = np.linspace(-np.pi,np.pi)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1,lw=2,color='red',label='sin(x)')
plt.plot(x,y2,lw=2,color='green',label='cos(x)')

plt.show()
```

**Multiplr plot in a single fig**

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(6,6))
x = np.linspace(-2,2,100)
plt.subplot(2,2,1)
y = x
plt.plot(x,y,color='red')
plt.xlabel('x')
plt.ylabel('x')

plt.subplot(2,2,2)
y = x**2
plt.plot(x,y,color='blue')
plt.xlabel('x')
plt.ylabel('x^2')

plt.subplot(2,2,3)
y = x**3
plt.plot(x,y,color='green')
plt.xlabel('x')
plt.ylabel('x^3')

plt.subplot(2,2,4)
y = x**4
plt.plot(x,y,color='black')
plt.xlabel('x')
plt.ylabel('x^4')

plt.suptitle('Functions')
plt.tight_layout()
plt.subplots_adjust(top=0.90)
```

# Three dimensional plotting

```python
from mpl_toolkits import mplot3d
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```python
#line plot

fig = plt.figure()
ax = plt.axes(projection='3d')

z = np.linspace(0,1,100)
x = z * np.sin(20*z)
y = z* np.cos(20*z)
ax.plot3D(x,y,z,'blue')
ax.set_title('3D line plot')
plt.show()
```

```python
#scatter plot

fig = plt.figure()
ax = plt.axes(projection='3d')

z = np.linspace(0,1,100)
x = z * np.sin(20*z)
y = z* np.cos(20*z)
c = x+y
ax.scatter(x,y,z,c =c)
ax.set_title('3D scatter plot')
plt.show()
```

```python
# contour plot

def f(x,y):
    return np.sin(np.sqrt(x ** 2 + y **2))

x = np.linspace(-6,6,30)
y = np.linspace(-6,6,30)
X , Y = np.meshgrid(x,y)
Z = f(X,Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X,Y,Z,50,cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3d contour')
plt.show()
```

```python
#view

def f(x,y):
    return np.sin(np.sqrt(x ** 2 + y **2))

x = np.linspace(-6,6,30)
y = np.linspace(-6,6,30)
X , Y = np.meshgrid(x,y)
Z = f(X,Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X,Y,Z,50,cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3d contour')
ax.view_init(60,35)
```

```python
#wireframe

def f(x,y):
    return np.sin(np.sqrt(x ** 2 + y **2))

x = np.linspace(-6,6,30)
y = np.linspace(-6,6,30)
X , Y = np.meshgrid(x,y)
Z = f(X,Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X,Y,Z,color='black')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('wireframe')
plt.show()
```

```python
#3d surface plot

x = np.outer(np.linspace(-2,2,30),np.ones(30))
y = x.copy().T
z = np.cos(x**2+y**2)

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x,y,z,cmap='viridis', edgecolor = 'none')
ax.set_title('Surface plot')
plt.show()
```
