```python
from plotnine import * 
import pandas as pd
```

```python
data = pd.read_csv('london2018.csv')
data.head(5)
```

```python
ggplot(data,aes('Month','Tmax')) + geom_line()
```

```python
mon = data.Month
ggplot(data,aes('Month','Tmax')) + geom_line() + scale_x_continuous(breaks=mon)
```

```python
mon_lab = ('jan','feb','mar','april','may','jun','july','agu','sep','oct','nov','dec')
ggplot(data,aes('Month','Tmax')) + geom_line(color='red') + scale_x_continuous(breaks=mon,labels=mon_lab)
```

```python
mon_lab = ('jan','feb','mar','april','may','jun','july','agu','sep','oct','nov','dec')
ggplot(data,aes('Month','Tmax')) + geom_col(color='red') + scale_x_continuous(breaks=mon,labels=mon_lab)
```

```python
tem = pd.melt(data,id_vars=['Month'],value_vars=['Tmax','Tmin'],var_name='Temp',value_name='DegC')
tem
```

```python
(ggplot(tem,aes('Month','DegC',fill='Temp')) + geom_col(position='dodge') + scale_x_continuous(breaks=mon))
```

```python
ggplot(tem,aes('Month','DegC',color='Temp')) + geom_line() + scale_x_continuous(breaks=mon)  
```

```python
ggplot(tem,aes('Month','DegC',color='Temp')) + geom_line() + scale_x_continuous(breaks=mon) + xlab('2018') + ylab('Temp in C') + ggtitle('Monthly max and min temp')
```

```python
tem1 = pd.melt(data,id_vars=['Month'],value_vars=['Tmax','Tmin','Rain','Sun'],var_name='Measure',value_name='Value')

ggplot(tem1,aes('Month','Value',fill='Measure')) + geom_col(show_legend=False) + scale_x_continuous(breaks=mon,labels=mon_lab) + facet_wrap('Measure',scales='free') + xlab('') + ylab('') + ggtitle('Weather data for london') + theme_light() + theme(panel_spacing=0.5,figure_size=(10,5))
```
