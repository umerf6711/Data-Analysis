```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('penguins')
sns.pairplot(df,hue='species')
plt.show()
```

```python
#load data set

tips = sns.load_dataset('tips')
tips.head(5)
```

```python
#plot data

import seaborn as sns
sns.set_style()
tips = sns.load_dataset('tips')
sns.relplot(x = "total_bill",y ='tip',col='time',hue='smoker',style='smoker',size='size', data=tips)
plt.show()
```

```python
#lmplot

sns.lmplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
sns.lmplot(x='size',y='tip',data=tips)
plt.show()
```

```python
sns.lmplot(x='size',y='tip',data=tips,x_estimator=np.mean)
plt.show()
```

```python
tips['big_tip'] = (tips.tip/tips.total_bill)>.15
sns.lmplot(x='total_bill',y='big_tip',y_jitter=.03,logistic=True,data=tips)
plt.show()
```

```python
#join plot
sns.jointplot(x='total_bill',y='tip',kind='reg',data=tips)
plt.show()
```

```python
dots = sns.load_dataset('dots')
sns.relplot(data=dots,kind='line',x='time',y='firing_rate',col='align',hue='choice',
                                           size='coherence',style='choice',facet_kws=dict(sharex=False))
plt.show()
```

```python
fmri = sns.load_dataset('fmri')
sns.relplot(data=fmri,kind='line',x='timepoint',y='signal',col='region',hue='event',style='event')
plt.show()
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker')
plt.show()
```

```python
#catplot for categorical data

sns.catplot(x='day',y='total_bill',hue='smoker',kind='swarm',data=tips)
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',hue='smoker',kind='violin',data=tips)
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',hue='smoker',kind='bar',data=tips)
plt.show()
```

```python
sns.relplot(data=df,x='bill_length_mm',y='bill_depth_mm',hue='body_mass_g')
plt.show()
```

```python
#grids

sns.set_style('whitegrid')
sns.stripplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
sns.set_style('white')
sns.stripplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
#despine

sns.set_style('white')
sns.stripplot(x='day',y='total_bill',data=tips)
sns.despine()
plt.show()
```

```python
sns.set_style('whitegrid')
sns.stripplot(x='day',y='total_bill',data=tips)
sns.despine(left=True,bottom=True)
plt.show()
```

```python
#scaling plot

sns.set_style('ticks')
sns.set_context('paper')
sns.stripplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
sns.set_style('ticks')
sns.set_context('poster')
sns.stripplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
#scaling fonts and linewidth

sns.set_style('ticks')
sns.set_context('poster',font_scale=.5,rc={'grig.linewidth':0.6})
sns.stripplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
sns.set_style('ticks')
sns.set_context('poster',font_scale=1,rc={'grig.linewidth':5})
sns.stripplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
#RC parameter

sns.set_style('ticks')
sns.set_context('poster')
sns.stripplot(x='day',y='total_bill',data=tips)
sns.plotting_context()
```

```python
#temporarily setting 

iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
with sns.axes_style('white'):
    plt.subplot(211)
    sns.swarmplot(x='species',y='petal_length',data=iris)
    plt.subplot(212)
    sns.violinplot(x='total_bill',data=tips)
    plt.show()
```

```python
#setting the figure size

fig , ax = plt.subplots(figsize=(10,4))
iris = sns.load_dataset('iris')
sns.swarmplot(x='species',y='petal_length',data=iris,ax=ax)
plt.show()
```

```python
titanic = sns.load_dataset('titanic')
g = sns.factorplot('class','survived','sex',data=titanic,kind='bar',size=6,aspect=2,palette='muted',legend=False)
plt.show()
```

```python
#heat map

np.random.seed(0)
sns.set()
data = np.random.rand(10,12)
ax = sns.heatmap(data,vmin=0,vmax=1)
plt.show()
```

```python
#heatmap color
df = pd.DataFrame(np.random.rand(50).reshape(10,5))
corr = df.corr()
ax1 = sns.heatmap(corr,cbar=0,linewidth=2,vmax=1,vmin=0,square=True,cmap='Blues')
plt.show()
```

```python
sns.set()
flights = sns.load_dataset('flights')
flights = flights.pivot('month','year','passengers')
ax = sns.heatmap(flights)
plt.title('Heatmap Flights Data')
plt.show()
```
