```python
import numpy as np
import pandas as pd
```

# Pandas Concatenation

```python
#df
one = pd.DataFrame({'id':list('12345'),
                    "sub_id": ['sub1','sub2','sub3','sub4','sub5'] ,
                    'name':['ali','ahmad','umer','usman','adil']})
two = pd.DataFrame({'id': list('12345'),
                    "sub_id": ['sub6','sub4','sub3','sub8','sub5'] ,
                    'name':['rom','shan','arsal','zain','hamza']})
print(one)
print('\n')
print(two)
```

```python
#concat()

print(pd.concat([one,two]))
```

```python
#using key

print(pd.concat([one,two],keys=['x','y']))
```

```python
#ignore_index=True : to avoid repeat index

print(pd.concat([one,two],keys=['x','y'],ignore_index=True))
```

```python
#axis = 1

print(pd.concat([one,two],axis=1))
```

**concat using append**

```python
print(one.append(two))
```

```python
print(one.append([two,one,two]))
```

# Data Functionality and time delta

```python
# range of dates

print(pd.date_range('1/12/2000',periods = 8))
```

```python
# range of dates, with freq

print(pd.date_range('1/12/2000',periods = 8,freq='M'))
```

```python
#bdate_range

print(pd.bdate_range('1/12/2000',periods = 8))
```

**Timedelta**

```python
# string

print(pd.Timedelta('2 days 2 hours 15 minutes 30 seconds'))
```

```python
#integer

print(pd.Timedelta(6,unit='m'))
```

```python
#Data Offset

print(pd.Timedelta(days=2))
```

**operations**

```python
s =pd.Series(pd.date_range('2012-1-1',periods=5,freq='D'))
tb = pd.Series([pd.Timedelta(days=i) for i in range(5)])
df = pd.DataFrame(dict(A=s,B=tb))
print(df)
```

```python
df.C = df.A + df.B
print(df.C)
```

# CategoricalData

```python
#category

s = pd.Series(list('abcdef'),dtype='category')
print(s)
```

```python
#pd.Categorical

cat = pd.Categorical(list('abcdefg'))
print(cat)
```

```python
cat = pd.Categorical(list('abcdefg'),list('abc'))
print(cat)
```

```python
#describe

cat = pd.Categorical(['a','b','c',np.nan],categories=['b','a','c'])
df = pd.DataFrame({'cat':cat,"s":['a','c','c',np.nan]})
print(df.describe())
print(df.cat.describe())
```

```python
#properties of category

print(cat.categories)
```

```python
print(cat.ordered)
```

```python
#append new cattegory

s = pd.Series(list('abcdef'),dtype='category')
s = s.cat.add_categories([4])
print(s.cat.categories)
```

```python
# removing 

print(s.cat.remove_categories('a'))
```
